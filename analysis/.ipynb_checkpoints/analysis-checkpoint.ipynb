```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
```

```python
def load_xvg_file(filename):
    x, y = [], []
    with open(filename) as f:
        for line in f:
            cols = line.split()
            if len(cols)==2:
                x.append(cols[0])
                y.append(cols[1])
    return np.array(x[1:],dtype=float),np.array(y[1:],dtype=float) #skip the first value (header)
```

```python
dt= 0.001
path = '/Users/fivos/Documents/Projects/2020/02_protein_dynamic_transition/02_simulations/mnt/Fivos/21_4aki_loop/data/'
```

### NPT

```python
plt.figure(figsize=[7,3])

plt.subplot(1,2,1)
for r in range(4):
    filename = "R%d/potential.xvg"%r
    x,y = load_xvg_file(path+filename)
    plt.plot(x[10:]*dt,y[10:],lw=1.5,label=r)
plt.ylabel('Potential Energy (KJ/mol)')
plt.xlabel('time (ns)')
plt.legend()
#plt.axhline(y=y[-1],ls='--',color='blue')

plt.subplot(1,2,2)
for r in range(4):
    filename = "R%d/volume.xvg"%r
    x,y = load_xvg_file(path+filename)
    plt.plot(x[10:]*dt,y[10:],lw=1.5)
plt.ylabel(r'Volume (nm$^3$)')
plt.xlabel('time (ns)')
plt.ylim(0,500)
#plt.axhline(y=y[-1],ls='--',color='green')

plt.tight_layout()
plt.savefig('/Users/fivos/Desktop/fig.png',dpi=300)
plt.show()
```

### X-ray scattering intensity

```python
plt.figure(figsize = [7,3.2])
i=12

i1 = 3
i2 = 11
T_ = np.arange(180,310,10)
Iq = np.zeros([len(T_),44])
colors = plt.cm.jet(np.linspace(0,1,len(T_)+1))

plt.subplot(1,2,1)
for T in np.arange(300,200,-10):
    filename = "R2/saxs_%d.xvg"%T_[i]
    q,Iq[i] = load_xvg_file(path+filename)
    if i==0 or i==4 or i==8 or i==12:
        plt.plot(q/10,Iq[i],'-',lw=1,c=colors[i],label='%dK'%T_[i],markersize=3)
    else:
        plt.plot(q/10,Iq[i],'-',lw=1,c=colors[i],markersize=3)
    i-=1
plt.axvline(x=q[i1]/10,ls='--',c='black',alpha=0.5)
plt.axvline(x=q[i2]/10,ls='--',c='black',alpha=0.5)

print(q[i1]/10,q[i2]/10)

plt.xlabel(r' Q $\rm[\AA^{-1}$]',fontsize=14)
plt.ylabel('I(Q)',fontsize=14)
plt.ylim(0,30)
plt.xlim(0,3)
plt.legend(frameon=False,loc='upper right', prop={'size': 10})

plt.subplot(1,2,2)
plt.plot(T_,Iq[:,i1]/Iq[:,i2],'o-',c='blue')
plt.ylabel(r'$\rm I_1/I_2$',fontsize=14)
plt.xlabel(r'T [K]',fontsize=14)
plt.xticks(np.arange(180,310,20))
plt.tight_layout()

plt.savefig('/Users/fivos/Desktop/fig_Iq0.png',dpi=300)
plt.show()
#plt.axhline(y=y[-1],ls='--',color='blue')
```
